```python
# Заполняем пропуски
# категориальные
df_clean = df_clean.fillna({
    "sex": "Unknown",
    "view": "Unknown",
    "finding": "Unknown"
})

# возраст: медианой (так как устойчивее среднего)
median_age = df_clean.approxQuantile("age", [0.5], 0.01)[0] if "age" in df_clean.columns else None
if median_age is not None:
    df_clean = df_clean.fillna({"age": median_age})

# Унификация диагнозов - новая колонка finding_clean
f = lower(col("finding"))

df_clean = df_clean.withColumn(
    "finding_clean",
    when(f.like("%covid%"), "COVID-19")
    .when(f.like("%sars%"), "COVID-19")
    .when(f.like("%pneumonia%"), "Pneumonia")
    .when(f.like("%normal%"), "Normal")
    .otherwise("Other")
)
```

Анализ категориальных признаков

В ходе анализа качества категориальных признаков были рассмотрены поля sex и view. Для признака sex были выявлены допустимые категории M, F и Unknown, что отражает неполноту метаданных для части пациентов и не является аномалией.

Для признака view было обнаружено несколько вариантов обозначения одной и той же проекции (например, AP, AP Supine, AP Erect). Для повышения качества аналитики и упрощения последующих SQL-запросов данные значения были унифицированы в обобщённые категории проекций.

```python
#Подготовка таблицы

from pyspark.sql.functions import col, lower, upper, when

df_sql = df_prep

# sex_clean
df_sql = df_sql.withColumn(
    "sex_clean",
    when(lower(col("sex")).isin("m","male"), "M")
    .when(lower(col("sex")).isin("f","female"), "F")
    .otherwise("Unknown")
)

# view_clean
df_sql = df_sql.withColumn(
    "view_clean",
    when(upper(col("view")).like("AP%"), "AP")
    .when(upper(col("view")) == "PA", "PA")
    .when(upper(col("view")) == "L", "L")
    .otherwise("Other")
)

df_sql.createOrReplaceTempView("xray")
```

```python
from pyspark.sql.functions import col, trim, when, concat, lit, coalesce, expr

# Нормализуем даты
df_sql2 = df_sql.withColumn("date_str", trim(col("date"))) \
    .withColumn(
        "date_norm",
        when(expr("date_str RLIKE '^[0-9]{4}$'"), concat(col("date_str"), lit("-01-01")))
        .otherwise(col("date_str"))
    ) \
    .withColumn(
        "date_parsed",
        coalesce(
            expr("try_to_timestamp(date_norm, 'MMMM d, yyyy')"),
            expr("try_to_timestamp(date_norm, 'MMMM dd, yyyy')"),
            expr("try_to_timestamp(date_norm, 'yyyy-MM-dd')")
        ).cast("date")
    )

df_sql2.createOrReplaceTempView("xray2")
```

Основные сложности при работе с данными были связаны с неполнотой и неоднородностью метаданных, включая пропуски, вариативность форматов дат и различия в обозначениях диагнозов и проекций. Эти проблемы были успешно устранены на этапе предобработки данных.

```python
from pyspark.sql.functions import udf, col
from pyspark.sql.types import StringType

# 1) UDF: возрастные группы
def age_to_group(age):
    if age is None:
        return "Unknown"
    try:
        age = int(age)
    except:
        return "Unknown"
    if age < 30:
        return "0-29"
    elif age < 50:
        return "30-49"
    elif age < 70:
        return "50-69"
    else:
        return "70+"

age_group_udf = udf(age_to_group, StringType())

df_udf = df_prep.withColumn("age_group_udf", age_group_udf(col("age_num")))

# 2) UDF: унификация диагнозов (на всякий случай именно UDF-ом)
def normalize_finding(finding):
    if finding is None:
        return "Unknown"
    f = str(finding).lower()
    if "covid" in f:
        return "COVID-19"
    if "pneumonia" in f:
        return "Pneumonia"
    if "normal" in f:
        return "Normal"
    if "sars" in f:
        return "SARS"
    return "Other"

finding_udf = udf(normalize_finding, StringType())

df_udf = df_udf.withColumn("finding_udf", finding_udf(col("finding")))

# 3) Фильтрация по критериям (пример: только COVID-19, взрослые 30+, валидная дата)
df_filtered = df_udf.filter(
    (col("finding_udf") == "COVID-19") &
    (col("age_num") >= 30) &
    (col("view").isNotNull())
)

print("All rows:", df_udf.count())
print("Filtered rows:", df_filtered.count())

# 4) Сохранение в Parquet (оптимизированный формат)
df_udf.write.mode("overwrite").parquet("/content/processed_full.parquet")
df_filtered.write.mode("overwrite").parquet("/content/processed_filtered.parquet")

print("Saved:")
print("/content/processed_full.parquet")
print("/content/processed_filtered.parquet")
```

В рамках обработки в PySpark были реализованы UDF-функции для категоризации возраста и нормализации диагнозов. Далее выполнена фильтрация данных по заданным критериям (например, COVID-19 и возраст ≥ 30 лет). Итоговые датасеты сохранены в формате Parquet, который является оптимизированным для хранения и последующей обработки в Spark.
